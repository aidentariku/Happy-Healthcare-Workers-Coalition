```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

#import study file:

burn_out_csv_path = "./Burn Out Questionnaire cleaned.csv"

burn_out_df = pd.read_csv(burn_out_csv_path)

burn_out_df.head()
```

```python
%pwd
```

```python
# Create a bar graph comparing Years of Service vs Overall Average Score 

categories= burn_out_df["Years of service?"]
values= burn_out_df["Overall Average"]



plt.figure(figsize=(10,7))
plt.bar(categories, values, color="blue")

plt.title("Years of Service vs Overall Average Score")
plt.xlabel("Years of Service")
plt.ylabel("Overall Average Score")
plt.grid(True)
```

```python
values = burn_out_df["Years of service?"]
categories = burn_out_df["Overall Average"]



plt.figure(figsize=(10,7))
plt.scatter(categories, values, color="blue", marker="o")

plt.title("Years of Service vs Overall Average Score")
plt.ylabel("Years of Service")
plt.xlabel("Overall Average Score")
plt.grid(True)
```
