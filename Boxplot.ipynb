```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('BurnOut.csv')
```

```python
# Assuming 'designation' is the correct column name and 'average_score' is the correct column name for 'BV'
designations = ['Nurse', 'Doctor', 'Radiographers', 'Physiotherapist', 'Supportive Staff', 'Minor Staff', 'Bio medical engineering']
df_filtered = df[df['Title'].isin(designations)]
```

```python
# Plotting the boxplot
plt.figure(figsize=(15, 7))
sns.boxplot(x=df_filtered['Title'], y=df_filtered['Overall Average '])
plt.title('Box plot of Overall Average Score by Designation')
plt.xlabel('Title')
plt.ylabel('Overall Average Score')
plt.xticks(rotation=45)
 # Set y-axis limits
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```
